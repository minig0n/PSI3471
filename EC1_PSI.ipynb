```python
import pandas as pd
cars_data = pd.read_csv("drive/MyDrive/Psi/vehicles_cleaned_train.csv")
```

```python
cars_data.head()
```

```python
# HIPÓTESES INICIAIS
# Variáveis que talvez possam ser descartadas:
# - Número de cilindros
# - Tipo de combustível
# - Tamanho
```

```python
# Implement dummies

cars_data_dummies = pd.get_dummies(cars_data, columns = ['condition', 'cylinders', 'fuel', 'transmission', 'size', 'type'])
```

```python
cars_data_dummies.head()
```

```python
# Append constant value column

cars_data_dummies_bias = cars_data_dummies
cars_data_dummies_bias['Constant'] = pd.Series([1 for x in range(len(cars_data_dummies_bias.index))])
```
